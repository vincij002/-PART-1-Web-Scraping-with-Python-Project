```python
# CREATING PREDICTORS FOR MACHINE LEARNING 

#Lets create a few basic predictors that we can use to create just a basic straightforward machine learning model. 
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes

#  this will convert it to a categorical data type in
# pandas because there's only two unique values in the column and then we'll convert it
# into integers using dot cat.codes so basically what this is doing is
#converting it from a string into categories and then converting those categories into numbers
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes

# we'll create a unique code for each opponent squad so what we'll say is matches opponent dot as type category dot cat dot codes 
# and this will do the same thing we just did with the venue column except with the opponent column 
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")

# The next column we'll look at is the hour column so maybe certain teams play
# better at certain times of day so what we'll do is we'll look at matches time,
# what we're going to replace and why so let's look at the time column so
# let's type out matches you can see the time column here and
# it's the hour colon the minute so what we basically want to do is remove the colon and the minutes and just keep
# the hour so we're going to do that with a string replacement so basically we're going to write a
# regular expression and this regular expression is basically going to replace the colon
# and the minutes with nothing 
# so that's what this replace method is doing is converting the hour to an integer because
# if you'll remember we need to be able to use a number as input to a machine learning algorithm

```

```python
matches["day_code"] = matches["date"].dt.dayofweek

# all right so there's one more column we're going to create before we build our initial model
# so this column is going to be called day code so what this will give us is a
# number for each day of the week so monday is a zero tuesday is one
# wednesday is two and so on. so there might be some useful information in there for the algorithm
```

```python
# CREATING OUR INITIAL MACHINE LEARNING MODEL

# Random Forst classifier is a type of machine learning model that can pick up non linearities in the data 

from sklearn.ensemble import RandomForestClassifier 
```

```python
# what we do is we just call the random forest classifier class and we'll pass in some parameters
# so n estimators is the number of individual decision trees we want to train so a random
# forest is a series of decision trees but each decision tree has slightly
# different parameters if you don't know about decision trees or random forests don't worry too much
# about it the higher this number is the longer the algorithm will take to
# run but potentially the more accurate it will be okay then we'll specify min samples
# split so this is basically the number of samples we want to have in a leaf of the
# decision tree before we actually split the node and basically the higher this is the less likely we are to overfit but
# the lower our accuracy on the training data will be so you might want to experiment a little bit with this value
# to find the the optimal one i'm just going to set it to 10 and then we want to set a random state
# so a random forest has a lot of random parameters in it so if we set a random state it means that if we run the random
# forest multiple times we'll get the same results as long as the data is the same

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
# Next we're going to split up our training and test data, this data is time series data so we need to be really careful when we
# split it to ensure that all the data in the test set comes after all of the data in the training set because in the real world
# you can't train an algorithm on data in the future and use it to predict the data in the past it has to be the other way around.
# You can only predict the future using the past so we'll basically take anything before 2025. So all the matches before 2025, and put
# those in our training set. 

train = matches[matches["date"] < '2024-1-1']
```

```python
# Then our test set will be anything in 2025. And that what the code below is doing. It's just subnetting our match data frame.

test = matches[matches["date"] > '2024-01-01']
```

```python
# why we split up into the train and test set the reason we do this is if you evaluate an algorithm if you see the error and
# how well it performed on the same data set that you trained it on your algorithm is going to look awesome
# this is kind of like having an open book test. you read the book and it feels like you know everything but then
# when you go to do something outside of the test maybe you don't actually know it so well because you were just reading
# the book. all the answers were in the book for the test so we want the algorithm to actually do
# well predicting future matches which is why we test it out on data that it hasn't been trained on because that
# enables us to see does this algorithm do well in making predictions on new data which is what we care about.
```

```python
# Then we'll create our predictors so this is just going to be a list of the predictor columns that we created so venue code, oppcode,
# hour, and day code.

predictors = ["venue_code", "opp_code", "hour", "day_code"] # These are our predictors. 
```

```python
# now we can do is we can actually fit our random forest model so we'll fit it using train predictors and we'll try to predict
# our target so the .fit method is basically going to train a random forest model with these predictors trying to predict this 
# target and the target is zero if the team lost or drew and a one if the team won.

rf.fit(train[predictors], train["target"])
```

```python
# now we can generate predictions using the predict method and we'll pass in our test data and predictors here.

preds = rf.predict(test[predictors])
```

```python
# Next we need to actually figure out how we determine the accuracy of the model
# and this is actually a really important choice so we're going to try a couple of different metrics and see which one
# makes more sense so what we'll do is we'll import from
# scikitlearn.metrics will import the accuracy score so accuracy score is a metric that will
# basically say if you predicted a win what percentage of the time did the team
# actually win if you predicted a loss what percentage of the time did the team actually lose and it'll combine that so
# basically looking at what percentage of the time was your prediction accurate

from sklearn.metrics import accuracy_score 
```

```python
# okay so what we'll do is we'll say accuracy equals accuracy score
# test target and then let's take a look at the
# then predictions i needed to pass those in so we pass in
# our actuals and then our predictions and then we can take a look at our accuracy

acc = accuracy_score(test["target"], preds)
```

```python
# so we can see our accuracy looks pretty good so when we predicted something would happen
# 60 % of the time that thing happened let's dig a little bit deeper into this
# though and see in which situations our accuracy was high versus low
# and in order to do that we first need to create a data frame and this data frame will combine our actual values and our predicted
# values so our actual is just the target and our prediction is the predictions.

combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
```

```python
# and then what we can do is actually create a crosstab using pandas so this is just a
# two-way table that will show us when we predicted a zero or a one what actually happened
# so what we'll do is call index pass in the index parameter which will be the rows of this cross tab
# which will just be the actual and then our columns will be our predictions

pd.crosstab(index=combined["actual"], columns=combined["prediction"])

# After running the above code you can see our columns are
# our predictions so when we predicted a loss or a draw you can see that most of
# the time we were correct 382 times we were correct, only 196 times we were wrong where the actual was a win
# but when we predicted a win we were actually wrong more often than we were right
# so since what we care about with this model is predicting wins we need to go back and revise our accuracy metric all little bit. 
```

```python
# so we're actually going to use a slightly different metric we're actually going to use something called the precision score 
# and what this is going to tell us is when we predicted a win what percentage of the time did the team actually win.

from sklearn.metrics import precision_score

# so what we'll do next is we'll again pass in the actual value so test target and then we'll pass in our predictions

precision_score(test["target"], preds)
```

```python
# all right and you can see here that our precision was only 40 so when we predicted a win the team only
# actually won 40 percent of the time which is not great precision so we can improve this model quite a bit 
# and we will do that in the next step
```

```python
# IMPROVING PRECISION WITH ROLLING AVERAGES

# so next we're going to create some more predictors to improve the accuracy of
# our model and to do this we're going to need to split our matches data frame up by team
# because what we want to do is compute rolling averages for team performance how did this team do in the past few
# games how many shots did they have how many goals were scored against them etc
# so what we're going to do is we're going to create something called grouped matches
# which is just going to be matches.groupedby. team so what this will do is it'll basically
# create one data frame for every squad in our data

grouped_matches = matches.groupby("team")
```

```python
# Next what we can do is we can actually look at a single group so we can do is group = grouped_matches
# .get_group("Atalanta") so what this will do is give us a single group from the data and this is just 
# Atalanta so these are all of the matches that Atalanta played in all. so what we want to do is compute 
# rolling averages so what we want to say is if we're at match week 4 how did Atalanta do in the previous three match
# weeks and then we can pass that into the algorithm and the algorithm can use that information to make better predictions
# about how they'll do in match week 4. for example if they lost all three of their games before match week four
# you might expect them to also lose match week four because they haven't been playing very well 

group = grouped_matches.get_group("Atalanta").sort_values("date")
```

```python
# So we're going to create a function called rolling averages and what this
# will do is take a group in it'll take a set of columns that we want to compute rolling averages for and then
# it'll take in a set of new columns that we want to assign the rolling averages to so the first thing we'll do
# is we'll sort our group by date because we want it to be sorted in ascending order of date because we want to look at
# the last three matches a team played and what their performance was like then
# we're going to create a variable called rolling stats and all we're going to do is take a set
# of columns which will be in this list calls that we pass in and compute the rolling averages for
# those columns and then this parameter is very important closed equals left
# so why is this important if you don't pass this in what pandas will do is compute the rolling averages for these
# three weeks and then assign that average to week three the problem with that is it uses
# knowledge of the future let's say we're trying to predict what happens in week three this rolling average method will
# actually include some of that into our week three data
# because it'll take the rolling average of these three weeks and assign it to week three what we instead want to have happen is
# if we're in week four we only want the rolling averages for the previous three weeks we don't want the week four data
# to be included in that so that's what this closed equals left does it basically says
# take the current week out and and ignore it when you're creating these rolling averages
# and then we'll just take the mean and then what we're going to do is we're going to assign these rolling stats back
# to our original data frame but with new names and i'll show you what calls and
# new calls will be in just a second and then finally what we need to do is we need to drop any missing values
# and let's talk about why we need to do that so let's say it's week two and you're trying to compute what happened
# in the previous three weeks you can't because there aren't three previous weeks of data to work with so what
# pandas will do here is it'll just put missing values in for all of all of
# those columns that we're computing rolling averages for so what this drop n a does is it removes
# all of the rows that have those missing values because you can't pass well many machine learning algorithms
# you can't pass missing values into some can deal with missing values but most can't and then we'll just return group

def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
# Next, let's define our columns list so these are the columns that we want to compute rolling averages for so we'll do cols = ["gf", "ga", "sh",
# "sot", "dist", "fk", "pk", "pkatt"] 

cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]

# Then we will create a new columns list which is just going to be a format string and 
# we're basically going to just add rolling on to the end of the column names

new_cols = [f"{c}_rolling" for c in cols]

```

```python
new_cols

# so you can see columns are the names of the original columns and new columns are the new columns we want to
# create with the rolling averages in them and you can see we just added underscore rolling to each column name
```

```python
# Okay now we can call rolling averages for a single group and our group is atalanta so this is just for the atalanta games rolling
# averages and pluralization always gets me okay so you can see we've now added in
# all of these extra columns that have information about what happened in the previous three matches
# so this is something we can pass into our machine learning algorithm and it will improve accuracy

rolling_averages(group, cols, new_cols)
```

```python
 # what we can do now is apply this to all of our matches so what we'll say is matches rolling equals matches.group by team this 
 # is what we did before except instead of calling get group we're going to call apply and what this will do is apply one function 
 # to each to each team essentially and that function is going to be this rolling averages function
 # and we'll pass in our columns and our new columns

matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
# What we did was we took our original matches data frame we grouped it by team which as i
# mentioned before creates one data frame for each team in our data and then we applied a function to each of those team
# data frames to compute the rolling averages so you can see we now have rolling
# averages for every match which is exactly what we need we do if you've noticed have this on the left though we have each team
# name as a separate level to our pandas index we actually don't need that so we'll go ahead and drop that extra index level it just
# makes it a little bit harder to work with the data frame so we'll call drop level team which will drop that extra index level

matches_rolling = matches_rolling.droplevel('team')
```

```python
 # Now let's take a look at matches rolling and see what happened

matches_rolling

# you can see we now have just a normal index on the left side and if you're not
# familiar with pandas indices, indices are how you call a specific row in pandas
# so each each row has a has an index that you use to call that row if you have a multi-level index like 2 cells above you have to
# use two values to access each row which sometimes you need but in this case we didn't so we just drop these extra
# index levels you might notice one other thing about our index we have 1207 rows but our index doesn't
# go that high which means that a lot of values in our index are being repeated but we actually want unique values in our index, 
```

```python
# so what we're going to do is just call matches_rolling.index = range(matches_rolling.shape [0]), and what this will do is 
# just assign values from 0 to 1207 to be our new in indices and you can see we now have new indices this is important just because 
# we want unique values for each index and this gives us that 


matches_rolling.index = range(matches_rolling.shape[0])
```

```python
# RETRAINING OUR MACHINE LEARNING MODEL 

# so now we have some new predictors Retraining our machine learning model 
# and we can make a new set of predictions using these predictors so we're going to create a function called make
# predictions this will just make it easy for us to basically continue to iterate on our algorithm because we don't have 
# to keep repeating the train test etc code so i'm going to copy and paste some code we already wrote so the first thing 
# we'll do is we'll split our data into a training set and a test set same thing we did before then we'll fit our random force model
# we did this before then we will create our predictions then we'll combine our predictions and our actuals together 
# then we'll calculate our our precision and then finally we'll return our precision value so we already did all of this we just
# put it into a single function so it's just a little bit easier to use 

def make_predictions(data, predictors):
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
# then what we could do is just call this function and what we'll do is we'll pass in our predictors so this is the original set of 
# four columns that we used and then we'll add in our new predictors as well which is in new calls

combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
# We can take a look at combined if we wanted to dig in a little bit more and see where our mispredictions were
# but combined doesn't tell us about which team actually played in each match so we
# can't see if we're mispredicting any specific team we can fix this by actually adding in
# some of that information to combine

combined
```

```python
# so we can do is call combine.merge and what we'll do is we'll match merge in some columns from matches rolling so emerging the date
# the team the opponent and the result and then what we'll do is we'll merge based on the index and that's what left index equals true
#right index equals true


combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)


# Basically what what pandas will do here is it'll look in our combined data frame it'll say okay the index for this row is 15
# it'll find the corresponding index in matches rolling and it'll merge the row um based on that 
```

```python
# COMBINING HOME AND AWAY PREDICTIONS

merged = combined.merge(combined, left_on=["date", "team"], right_on=["date", "opponent"])
```
